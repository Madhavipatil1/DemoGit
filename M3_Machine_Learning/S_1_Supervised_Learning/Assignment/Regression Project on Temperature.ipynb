## <span style="color:red; font-weight:bold;">üîß Step 1: Importing Libraries üìö</span>
### First, let's bring in some important libraries that will help us with data handling and creating a predictive model! We'll use:
- **pandas** to handle our dataset.
- **matplotlib** to visualize the data with cool graphs. üìä
- **sklearn** to perform Linear Regression (don't worry, we'll explain this later!). ü§ì


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

## <span style="color:red; font-weight:bold;">üìÇ Step 2: Loading the Data üåç</span>
### Now, let's read our data using `pandas`. The data contains the year and average temperature. Each row represents a different year and its corresponding average temperature across the world. Let's see what our data looks like! üëÄ


```python
data= pd.read_csv("temperature_data.csv")
```

```python
data
```

## <span style="color:red; font-weight:bold;">ü§ñ Step 3: Creating the Model üß†</span>
### We are now going to prepare our **Linear Regression** model. The independent variable here is the **Year** (X), and the dependent variable is the **Average Temperature** (y). In simple words:
- **X (Year)**: The year we want to predict the temperature for.
- **y (Temperature)**: The temperature in that year.

Year is chosen as the X-axis (independent variable) because it represents the time or progression of years, and it helps us predict future outcomes. The temperature is expected to change over time, so the year acts as the predictor.

Average Temperature is placed on the Y-axis (dependent variable) because it is what we are trying to predict. The temperature depends on the year, and we are modeling how temperature trends change over time. Hence, temperature is the result or outcome based on the year, making it the dependent variable.

```python
X=data['Year'].values.reshape(-1,1)
```

```python
y= data['AverageTemperature'].values
```

## Linear Regression

Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. It aims to find the best-fitting straight line that minimizes the difference between the predicted and actual values.

### Simple Linear Regression

In simple linear regression, there is one independent variable and the relationship is modeled with a straight line. The formula for simple linear regression is:


Linear regression is commonly used for predictive modeling, trend analysis, and understanding the impact of various factors.


## <span style="color:red; font-weight:bold;">üèãÔ∏è‚Äç‚ôÇÔ∏è Step 4: Training the Model üéØ</span>
### Now, let‚Äôs **train** our model! We are using the years and their respective average temperatures to make the model learn how to predict temperatures based on a given year.


```python
model = LinearRegression()
```

### `model.fit()` Method

The `model.fit()` method is used to train a machine learning model on a given dataset. It adjusts the model's parameters to minimize the error between predicted and actual values. 

**Key Points:**
- **Training**: Updates model parameters based on input data.
- **Optimization**: Minimizes the loss function using optimization algorithms.
- **Data Integration**: Uses training data to learn patterns and make predictions.
- **Validation**: Can include validation data to tune the model and prevent overfitting.


```python
model.fit(X,y)
```

## <span style="color:red; font-weight:bold;">üîÆ Step 5: Predicting Future Temperatures üöÄ</span>
### Our model is now trained! Let's use it to predict future temperatures from the year **2021 to 2030**. These will be our predicted future temperatures based on past data.


```python
future_years= np.array(range(2021,2031)).reshape(-1,1)
```

## <span style="color:red; font-weight:bold;">üìà Step 6: Analyzing the Trend üîç</span>
### The model will now predict the temperature for the upcoming years. Let's store these predictions in a variable and observe the changes in temperatures over time.


```python
predicted_temperatures= model.predict(future_years)
```

```python
predicted_temperatures
```

## <span style="color:red; font-weight:bold;">üé® Step 7: Visualizing the Data üìä</span>
### Let's plot the data! Here's what we're doing:
- **Blue dots** represent the actual data from the past.
- The **Red line** shows the regression line, which is the trend line based on past data.
- The **Green line** represents our future predictions from the year 2021 to 2030.
- Watch how the temperatures change over time! üå°Ô∏è


```python
plt.scatter(X,y, color='blue', label='Actual Data')
plt.plot(X,model.predict(X), color='red', label='Regression Line')
plt.plot(future_years,predicted_temperatures,color='green', label='Future Predictions')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title('Year vs Avg Temperature')
plt.legend()
plt.show()
```
