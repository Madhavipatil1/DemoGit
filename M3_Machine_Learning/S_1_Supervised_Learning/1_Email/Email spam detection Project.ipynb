<span style="color:red; font-size:24px">## Step 1: Importing Libraries</span>


```python
# Importing necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

### Explanation:

- **pandas**: Helps to handle data in rows and columns.
- **train_test_split**: Splits the dataset into training and testing sets.
- **CountVectorizer**: Converts text to numbers for the model.
- **MultinomialNB**: The machine learning model used for spam detection.
- **accuracy_score** and **confusion_matrix**: Evaluate the model‚Äôs performance.
- **seaborn** and **matplotlib**: For visualizing data.


<span style="color:red; font-size:24px">## Step 2: Loading the Dataset</span>

```python
# Load the dataset
df = pd.read_csv("email_spam_dataset.csv")

# Display the first few rows of the dataset
print("Dataset:")
print(df.head())
```

Explanation:

We are loading the dataset saved in CSV format into a pandas DataFrame using pd.read_csv().

Then we use df.head() to display the first 5 rows of the dataset.

<span style="color:red; font-size:24px">## Step 3: Data Preprocessing</span>

```python
# Convert labels (spam/not spam) to numerical format: 'spam' -> 1, 'not spam' -> 0
df['Label'] = df['Label'].map({'spam': 1, 'not spam': 0})

# Split the dataset into features (X) and labels (y)
X = df['Email']  # This is the email content
y = df['Label']  # This is whether it's spam (1) or not spam (0)

# Split the data into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display the size of training and testing sets
print(f"\nTraining size: {X_train.shape[0]}, Testing size: {X_test.shape[0]}")
```

### Explanation:

- **Convert Labels**: We convert the labels from text (**spam**/**not spam**) to numbers (**1** and **0**).

- **Split Data**: We split the data into:
  - **Features (X)**: The email content.
  - **Labels (y)**: Whether each email is spam (1) or not spam (0).

- **Divide Data**: Finally, the data is divided into:
  - **Training Set (70%)**: Used to teach the model.
  - **Testing Set (30%)**: Used to check how well the model works, using `train_test_split()`.


<span style="color:red; font-size:24px">## Step 4: Feature Extraction</span>

```python
# Use CountVectorizer to convert text emails to numerical values (bag of words)
vectorizer = CountVectorizer()

# Fit the vectorizer on training data and transform it
X_train_transformed = vectorizer.fit_transform(X_train)

# Only transform (not fit) the test data
X_test_transformed = vectorizer.transform(X_test)
```

### Explanation:

- **CountVectorizer**: We use `CountVectorizer` to convert the text data into numerical values.

- **Training Data**: The training data is both **fitted** and **transformed** to create numerical representations.

- **Test Data**: The test data is only **transformed** using the same conversion rules from the training data.


<span style="color:red; font-size:24px">## Step 5: Training the Model</span>

```python
# Train a Naive Bayes model using the training data
model = MultinomialNB()
model.fit(X_train_transformed, y_train)

# Print that the model training is complete
print("\nModel Training Complete!")
```

### Explanation:

- **Multinomial Naive Bayes Model**: We use this model to train the data on the training set.

- **Learning Patterns**: The model learns the patterns of spam and non-spam emails.


<span style="color:red; font-size:24px">## Step 6: Testing the Model</span>

```python
# Predict the results on the test data
y_pred = model.predict(X_test_transformed)

# Display predictions for the test set
print("\nPredictions on Test Set:")
print(y_pred)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of the Model: {accuracy * 100:.2f}%")
```

### Explanation:

- **Testing the Model**: The model is tested on the test set, and predictions are made to determine if the emails are spam or not.

- **Accuracy**: We calculate and display the accuracy of the model to see how well it performs.


<span style="color:red; font-size:24px">## Step 7: Extra Visualization - Actual vs Predicted Spam Count</span>

```python

# Visualization: Bar Plot for Actual vs Predicted Spam/Not Spam Counts
# Count the number of actual and predicted spam/not spam emails
actual_counts = pd.Series(y_test).value_counts()
predicted_counts = pd.Series(y_pred).value_counts()

# Create a bar plot to compare actual vs predicted counts
plt.figure(figsize=(8, 5))
bar_width = 0.35
index = range(len(actual_counts))

plt.bar(index, actual_counts, bar_width, label="Actual", color='blue')
plt.bar([i + bar_width for i in index], predicted_counts, bar_width, label="Predicted", color='green')

# Adding labels, title, and legend
plt.xlabel('Email Classification')
plt.ylabel('Count')
plt.title('Actual vs Predicted: Spam vs Not Spam')
plt.xticks([i + bar_width/2 for i in index], ['Not Spam', 'Spam'])
plt.legend()
plt.show()
```

### Visualization: Bar Plot for Actual vs Predicted Spam/Not Spam Counts

- **Counting Emails**:
  - **Actual Counts**: We count how many emails are actually **spam** and **not spam** in the test set.
  - **Predicted Counts**: We count how many emails the model predicted as **spam** and **not spam**.

- **Creating the Bar Plot**:
  - **Plot Setup**: We create a bar plot to compare the actual counts versus the predicted counts.
  - **Bar Width**: Each bar is set to a width of `0.35` to ensure they don‚Äôt overlap.
  - **Bar Colors**:
    - **Actual Counts**: Displayed in **blue**.
    - **Predicted Counts**: Displayed in **green**.
  - **X-axis**: Shows the email classifications (**Spam** and **Not Spam**).
  - **Y-axis**: Shows the count of emails for each classification.
  - **Title**: The plot is titled **"Actual vs Predicted: Spam vs Not Spam"**.
  - **Legend**: Helps to differentiate between actual and predicted counts.

- **Displaying the Plot**: Finally, the plot is shown to visualize how well the model‚Äôs predictions match the actual email classifications.


```python

# Interactive input for classification
def classify_email(email):
    email_transformed = vectorizer.transform([email])
    prediction = model.predict(email_transformed)[0]
    return 'Spam' if prediction else 'Not Spam'

print("\nüîç Enter your email below for classification (type 'exit' to quit):")
while True:
    user_input = input("Email: ")
    if user_input.lower() == 'exit':
        break
    result = classify_email(user_input)
    print(f"Classification: {result}")
```

### How Does the Spam Detector Work?

1. **Learning from Examples**: 
   The computer "learns" from a bunch of emails that are already labeled as **"Spam"** or **"Not Spam"**. It notices which words are common in spam emails and which are not.

2. **Changing Words to Numbers**: 
   The computer converts the words in each email into numbers, so it can understand them better.

3. **Making Predictions**:
   - When you give the computer a new email, it changes the words in that email to numbers too.
   - It then checks these numbers against what it learned from the examples.
   - If the numbers match patterns it learned from spam emails, it will say the email is **"Spam"**. If they match patterns from non-spam emails, it will say **"Not Spam"**.

### Example:
- If an email has words like  **"discount"**, which the computer learned are often in spam emails, it might classify it as **"Spam"**.
- If the email has more normal words that the computer learned are often in non-spam emails, it will classify it as **"Not Spam"**.
