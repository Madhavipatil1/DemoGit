# What is Regression?
Regression is a type of mathematical and statistical technique used to find relationships between different variables. Simply put, it helps us understand how one thing (like rainfall) affects another thing (like umbrella sales). In the world of machine learning and data science, we use regression to predict future outcomes based on past data.

# **Where Do We Use Regression?**

### Regression is used in many real-life situations to predict future trends or outcomes. Some examples include:

 - **Business:** Predicting future sales based on past data (like our umbrella sales example).
 - **Weather:** Predicting future temperatures or rainfall based on historical weather data.
 - **Finance:** Predicting stock prices or a company's future earnings.
 - **Healthcare:** Predicting a patient's risk of developing a disease based on their medical history.
 - **Education:** Predicting a student's grades based on study hours and other factors.



# **Understanding Dependent and Independent Variables in Linear Regression**

 - **Dependent Variable (Target Variable):** This is the variable you're trying to predict or explain. It's the outcome that depends on other factors. In a linear regression model, it's usually denoted by \( y \).

 - **Independent Variable (Predictor Variable):** This is the variable that you use to make predictions about the dependent variable. It's the factor that influences or predicts the outcome. In linear regression, it's typically denoted by \( x \).

 In essence, the dependent variable is what you want to understand or forecast, while the independent variables are the factors you believe influence or determine the outcome.




## Welcome to our project! üéâ Today, we will explore how rainfall affects the number of umbrellas sold using a simple linear regression model.


## <span style="color:red; font-weight:bold;">üîß Step 1: Importing Libraries üìö</span>
### First, let's bring in some important libraries that will help us with data handling and creating a predictive model! We'll use:
- **pandas** to handle our dataset.
- **matplotlib** to visualize the data with cool graphs. üìä
- **sklearn** to perform Linear Regression (don't worry, we'll explain this later!). ü§ì


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

## <span style="color:red; font-weight:bold;">üìÇ Step 2: Loading the Data üåç</span>
### Now, let's read our **Rainfall vs Umbrellas Sold** data using `pandas`. Each row in the dataset shows:
- **Rainfall (mm)**: The amount of rainfall recorded.
In this dataset, 
**Rainfall (mm)** is chosen as the X-axis (independent variable) because it is the factor that influences the outcome. As rainfall increases, the number of umbrellas sold typically changes, making it the cause or predictor.


- **Umbrellas Sold**: The number of umbrellas sold based on the amount of rain. ‚òÇÔ∏è
**Umbrellas Sold** is placed on the Y-axis (dependent variable) because it depends on the amount of rainfall. The more it rains, the more umbrellas people are likely to buy. Therefore, umbrella sales are the result or response to rainfall, making them the dependent variable.


```python
data = pd.read_csv("rainfall_data.csv")
```

```python
data
```

## <span style="color:red; font-weight:bold;">ü§ñ Step 3: Creating the Model üß†</span>
### We are now going to prepare our **Linear Regression** model. In this case:
- **X (Rainfall)**: The amount of rainfall.
- **y (Umbrellas Sold)**: The number of umbrellas sold when it rained.


```python
X = data['Rainfall'].values.reshape(-1,1)
```

```python
y = data['Umbrellas Sold'].values
```

## Linear Regression

Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. It aims to find the best-fitting straight line that minimizes the difference between the predicted and actual values.

## <span style="color:red; font-weight:bold;">üèãÔ∏è‚Äç‚ôÇÔ∏è Step 4: Training the Model üéØ</span>
### Now, let‚Äôs **train** our model! We will use the amount of **rainfall** and corresponding **umbrellas sold** to teach our model how these two are related.


```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

## <span style="color:red; font-weight:bold;">üîÆ Step 5: Predicting Umbrellas Sales for Future Rainfall üöÄ</span>
### Our model is now trained! Let's predict how many umbrellas would be sold for future rainfall amounts from **110mm to 130mm**.


```python
future_rainfall = np.array(range(110, 131)).reshape(-1, 1)
```

## <span style="color:red; font-weight:bold;">üìà Step 6: Analyzing the Trend üîç</span>
### The model will now predict the number of umbrellas that will be sold for different rainfall amounts. Let's store these predictions in a variable.


```python
predicted_sales = model.predict(future_rainfall)
```

```python
predicted_sales
```

## <span style="color:red; font-weight:bold;">üé® Step 7: Visualizing the Data üìä</span>
### Let's plot the data! Here's what we're doing:
- **Blue dots** represent the actual data of rainfall vs umbrellas sold.
- The **Red line** shows the regression line, indicating the trend between rainfall and umbrellas sold.
- The **Green line** represents our future predictions from **110mm to 130mm** rainfall.


```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.plot(future_rainfall, predicted_sales, color='green', label='Future Predictions')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Umbrellas Sold')
plt.title('Rainfall vs Umbrellas Sold')
plt.legend()
plt.show()
```
