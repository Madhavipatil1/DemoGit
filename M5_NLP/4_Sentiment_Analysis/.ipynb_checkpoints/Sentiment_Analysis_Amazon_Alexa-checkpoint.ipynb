# <span style="color:red;"> Try It Yourself: What Does the Model Think of Your Feedback? </span>


### Preprocessing and Modelling 
Before our system can predict if your review is positive or negative, it needs to understand the words in your review. Heres how we prepare your words for the model:

1. **Replace Characters:** First, we replace any non-alphabet characters (like numbers and punctuation) with a space. This helps us focus only on the words.
2. **Convert to Lowercase:** We then convert all the words to lowercase to avoid confusion (e.g., "Good" and "good" should be treated the same).
3. **Split Into Words:** Your review is split into individual words so the system can understand each part of your feedback.
4. **Remove Stopwords:** We remove common words that dont carry much meaning, like "is," "and," or "the."
5. **Stemming:** Finally, we take each word and reduce it to its root form. For example, "playing" becomes "play," so the model understands all variations of the word.

These steps help our model focus on the important parts of your review, making it easier to classify your feedback correctly!


# Code Explanation

This code is designed to analyze customer reviews and predict whether they are positive or negative using a machine learning model. Heres how it works step-by-step:

1. <font color="blue">**Importing Libraries**</font>: 
   - **pandas**: This library helps us manage and analyze data in a structured way (like a spreadsheet).
   - **re**: A library that allows us to work with text patterns, helping us clean the review text.
   - **nltk.corpus.stopwords**: Provides a list of common words (like "the", "is", "and") that we remove from the reviews because they don't add much meaning.
   - **nltk.stem.porter**: Used to reduce words to their root form (for example, "playing" becomes "play").
   - **sklearn.ensemble.RandomForestClassifier**: This is the machine learning model we will use to make predictions.
   - **sklearn.model_selection.train_test_split**: A tool to split our data into training and testing sets.
   - **sklearn.metrics.accuracy_score**: This helps us measure how well our model is performing.
   - **sklearn.feature_extraction.text.CountVectorizer**: Converts the cleaned text into a format that the model can understand.

2. <font color="blue">**Loading Data**</font>: 
   - We assume that the reviews have already been loaded into a DataFrame called `data`. This DataFrame contains a column with the actual reviews and another with feedback (positive or negative).

3. <font color="blue">**Defining the Preprocessing Function**</font>:
   - **preprocess_review(review)**: This function cleans up the review text. It does the following:
     - **Replace non-alphabet characters**: Any character that is not a letter is replaced with a space. This helps to keep only the words.
     - **Convert to lowercase**: Changes all letters to lowercase so that "Good" and "good" are treated the same.
     - **Split into words**: Breaks the review into individual words.
     - **Remove stopwords**: Filters out common words that do not add much meaning.
     - **Stemming**: Reduces words to their base form (root), which helps in treating different forms of a word as the same.

4. <font color="blue">**Cleaning the Reviews**</font>:
   - The `preprocess_review` function is applied to each review in the 'verified_reviews' column. The cleaned reviews are stored in a new column called 'cleaned_reviews'.

5. <font color="blue">**Vectorizing the Reviews**</font>:
   - **CountVectorizer()**: Initializes a tool that will convert our cleaned text into numbers that the model can understand. 
   - **fit_transform()**: This method learns the vocabulary from the cleaned reviews and creates a matrix of token counts.

6. <font color="blue">**Splitting the Data**</font>:
   - The data is divided into two parts: one for training the model (X_train, y_train) and one for testing its accuracy (X_test, y_test). This is done to ensure that we can evaluate how well the model performs on new, unseen data.

7. <font color="blue">**Training the Model**</font>:
   - A Random Forest model is created with 100 trees. This model learns from the training data (X_train, y_train).

8. <font color="blue">**Making Predictions**</font>:
   - The trained model predicts the feedback (positive or negative) for the reviews in the testing set (X_test).

9. <font color="blue">**Evaluating the Model**</font>:
   - **accuracy_score()**: This function compares the model's predictions with the actual feedback from the test set and calculates how many predictions were correct.
   - Finally, it prints out the accuracy of the model, showing how well it performed in classifying the reviews.


```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer

# Assuming you have already loaded the data into a DataFrame called 'data'
# Function to preprocess the reviews
def preprocess_review(review):
    # Replace non-alphabet characters with space
    review = re.sub('[^a-zA-Z]', ' ', review)
    # Convert to lowercase
    review = review.lower()
    # Split into words
    review = review.split()
    # Remove stopwords
    review = [word for word in review if word not in stopwords.words('english')]
    # Stemming
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review]
    return ' '.join(review)

# Apply the preprocessing function to the 'verified_reviews' column
data['cleaned_reviews'] = data['verified_reviews'].apply(preprocess_review)

# Initialize CountVectorizer
cv = CountVectorizer()
# Fit and transform the cleaned reviews
words = cv.fit_transform(data['cleaned_reviews'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(words, data['feedback'], test_size=0.2, random_state=42)

# Initialize and train the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"Model Accuracy: {accuracy}")
```

# <span style="color:red;"> Review Magic: Predicting Your Feedback! </span>


```python
def get_user_feedback():
    # Ask the user for their review
    user_review = input(" How was your review? ")

    # Preprocess the user's review
    cleaned_review = preprocess_review(user_review)

    # Transform the cleaned review into the same format used for training
    user_words = cv.transform([cleaned_review])

    # Make a prediction based on the processed review
    prediction = model.predict(user_words)

    # Provide feedback based on the prediction
    if prediction[0] == 1:  # Assuming 1 is positive
        print(" Thank you for your positive feedback! We're glad you liked it! ")
    else:
        print(" We're sorry to hear that! We'll work on improving. Thank you for your feedback!")

# Run the function to start interacting with the user
get_user_feedback()
```

## Explanation

* **Function Definition**: We define a function named `get_user_feedback()` to handle user interaction.
* **Input Prompt**: The program prompts the user to input their review with a friendly message and a star emoji.
* **Preprocessing**: The input review is cleaned and preprocessed using the previously defined `preprocess_review()` function.
* **Transformation**: The cleaned review is transformed into a format suitable for the model using `cv.transform()`.
* **Prediction**: The model predicts whether the review is positive or negative.
* **Feedback Response**:
  * If the prediction indicates positive feedback (assumed to be represented by `1`), the program prints a cheerful message with celebration and smiley emojis.
  * If the feedback is negative, it responds with a sympathetic message and a sad emoji, thanking the user for their input.
