## <span style="color: red;">5. Convert to Grayscale 🌈➡️⚫️</span>


In this cell, we will convert our colorful image to grayscale.

- We use `cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)` to perform the conversion. Here, `img` is our original color image, and `cv2.COLOR_BGR2GRAY` is the code that tells OpenCV to convert the image to grayscale.
- We store the result in a new variable called `gray`.
- Finally, we display the grayscale image using `plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))`. We convert it back to RGB for correct color representation in Matplotlib.

Grayscale images simplify the processing by reducing complexity, making it easier for our algorithms to work.


```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))

# Convert the original image ('img') to grayscale using OpenCV's cvtColor function and then display the grayscale image using Matplotlib.
```

## <span style="color: red;">6. Apply Filters and Edge Detection 🔍</span>

In this cell, we will apply a filter to reduce noise and then detect edges in the grayscale image.

- We first apply a **bilateral filter** with `cv2.bilateralFilter(gray, 11, 17, 17)`. This filter smooths the image while keeping edges sharp. The parameters control the filter's size and strength.
- We then use the **Canny edge detection** algorithm with `cv2.Canny(bfilter, 30, 200)`. This function detects the edges in the filtered image. The two numbers (30 and 200) are thresholds that determine how sensitive the edge detection should be.
- Finally, we display the edge-detected image using `plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))`.

By detecting edges, we can identify important shapes in the image that we will work with next.


```python
bfilter = cv2.bilateralFilter(gray, 11, 17, 17) 
edged = cv2.Canny(bfilter, 30, 200) 
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))

# Apply bilateral filtering for noise reduction and Canny edge detection on the grayscale image ('gray'), then display the result using Matplotlib.
```

```python
keypoints = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(keypoints)
contours = sorted(contours, key=cv2.contourArea, reverse=True)[:10]

# Find contours in the Canny edge-detected image ('edged'), extract and sort the top 10 contours based on area using OpenCV and imutils.
```


## <span style="color: red;">7. Find Contours 🔲</span>

In this cell, we will find the contours of shapes in the edged image.

- We use `cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)` to find the contours in the edge-detected image. This function returns the contours found in the image.
- We then grab the contours using `imutils.grab_contours(keypoints)`, which formats the data correctly.
- Next, we sort the contours by area using `sorted(contours, key=cv2.contourArea, reverse=True)[: 10]` to keep only the largest ones. This helps us focus on the most significant shapes.
- We initialize `location` as `None` and use a loop to check each contour. We approximate the polygon shape with `cv2.approxPolyDP(contour, 5, True)`. If the shape has four vertices (indicating a rectangle), we store its location and break out of the loop.

This step is crucial as we want to find the area in the image that we will analyze for text.


```python
location = None
for contour in contours:
    approx = cv2.approxPolyDP(contour, 5, True)
    if len(approx) == 4:
        location = approx
        break
        
# Iterate through the sorted contours, approximate the polygon shape, and if the contour has four vertices, store its location and break from the loop.
```


## <span style="color: red;">8. Create a Mask 🛡️</span>

In this cell, we will create a mask that highlights the area of interest.

- We create a blank mask using `np.zeros(gray.shape, np.uint8)`, which creates an array of zeros (black image) with the same dimensions as our grayscale image.
- We draw the identified contour on the mask with `cv2.drawContours(mask, [location], 0, 255, -1)`. This fills the area inside the contour with white (255) on the black mask.
- Finally, we display the new mask using `plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))`.

The mask will help us isolate the area where we want to read text.


```python
location
```

```python
mask = np.zeros(gray.shape, np.uint8)
new_image = cv2.drawContours(mask, [location], 0,255, -1)
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))

# Create a binary mask using the identified polygonal location and display the masked region in the original grayscale image using Matplotlib.
```

## <span style="color: red;">9. Apply Bitwise AND Operation ✨</span>

In this cell, we will keep only the part of the original image that corresponds to our mask.

- We use the `cv2.bitwise_and` function: `new_image = cv2.bitwise_and(img, img, mask=mask)`. This function takes the original image and the mask, applying a bitwise AND operation. It keeps only the area of the image where the mask is white (255).
- We then display this new image, which shows only the area of interest.

This operation effectively isolates the region in the image where we believe text is located.


```python
new_image = cv2.bitwise_and(img, img, mask=mask)
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
# Apply bitwise AND operation to overlay the masked region on the original image and display the result using Matplotlib.
```

## <span style="color: red;">10. Crop the Image ✂️</span>

In this cell, we will crop the image to focus on the area where the text is located.

- We find the coordinates of the white areas in the mask using `np.where(mask == 255)`, which gives us the x and y coordinates.
- We determine the minimum and maximum coordinates with `(x1, y1) = (np.min(x), np.min(y))` and `(x2, y2) = (np.max(x), np.max(y))`.
- Finally, we crop the grayscale image using `cropped_image = gray[x1:x2 + 1, y1:y2 + 1]`.

This gives us a smaller image that should contain just the text we want to read.


```python
(x,y) = np.where(mask==255)
(x1, y1) = (np.min(x), np.min(y))
(x2, y2) = (np.max(x), np.max(y))
cropped_image = gray[x1:x2+1, y1:y2+1]

# Find the coordinates of non-zero values in the mask, determine the bounding box, and extract the corresponding region from the original grayscale image.
```

## <span style="color: red;">11. Display the Cropped Image 🔍</span>

In this cell, we display the cropped region of the grayscale image.

- We use `plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))` to show the cropped image. This area should contain the text we want to recognize.

Displaying this cropped image allows us to verify that we have successfully isolated the text region.


```python
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
# Display the cropped region of the grayscale image using Matplotlib after converting it to RGB format.
```


## <span style="color: red;">12. Optical Character Recognition (OCR) 🧠</span>

In this cell, we will use EasyOCR to extract text from the cropped image.

- We create an EasyOCR reader configured for English with `reader = easyocr.Reader(['en'])`.
- We then call `reader.readtext(cropped_image)` to perform OCR on the cropped image, storing the result in the variable `result`.

This step is where the magic happens: EasyOCR analyzes the cropped image and extracts any text it finds!


```python
reader = easyocr.Reader(['en'])
result = reader.readtext(cropped_image)
result

# Use the EasyOCR library to perform optical character recognition (OCR) on the cropped image, reading and extracting text information.
```

## <span style="color: red;">13. Display Recognized Text 📝</span>

In this cell, we will overlay the recognized text onto the original image.

- We extract the recognized text from the result using `text = result[0][-2]`.
- We set the font and position where we want to place the text using `cv2.putText`. We


```python
text = result[0][-2]
font = cv2.FONT_HERSHEY_SIMPLEX
res = cv2.putText(img, text=text, org=(approx[0][0][0], approx[1][0][1]+60), fontFace=font, fontScale=1, color=(0,255,0), thickness=2, lineType=cv2.LINE_AA)
res = cv2.rectangle(img, tuple(approx[0][0]), tuple(approx[2][0]), (0,255,0),3)
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))

# Overlay the recognized text on the original image, draw a rectangle around the identified area, and display the final result using Matplotlib.
```
